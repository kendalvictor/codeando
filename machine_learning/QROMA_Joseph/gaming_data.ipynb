```python
#importaciones
import pandas as pd

#variables
sep = ','
ruta = 'data_qroma_train.csv'

data = pd.read_csv(ruta, encoding= "ISO-8859-1", sep=sep, low_memory=False)
data.head(50)
```

```python
def null_verificator(data):
    if not data.isnull().any().any():
        return "DATA LIMPIA DE NULOS"
    
    view_info = pd.DataFrame(
        pd.concat(
            [data.isnull().any(), 
             data.isnull().sum(),
             data.dtypes], 
            axis=1
        )
    )
    view_info.columns = ['Nulos', 'Cantidad', 'Tipo Col']
    return view_info

null_verificator(data)
```

```python
for col in ['NUM_PEDIDO', 'CODIGO_CONGLOMERADO_CLIENTE', 'CODIGO_CARACTERSTICA']:
    print("/////////////////// ", col)
    print(col, data[col].value_counts(dropna=False))
```
