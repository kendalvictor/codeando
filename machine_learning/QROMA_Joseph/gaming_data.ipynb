```python
#importaciones
import pandas as pd

#variables
sep = ','
ruta = 'data_qroma_train.csv'

data = pd.read_csv(ruta, encoding= "ISO-8859-1", sep=sep, low_memory=False)
data.head(10)
```

```python
def get_memory_usage(data):
    return '{} MB'.format(data.memory_usage(deep=True).sum() / 1024 ** 2)

get_memory_usage(data)
```

```python
def null_verificator(data):
    if not data.isnull().any().any():
        return "DATA LIMPIA DE NULOS"
    
    view_info = pd.DataFrame(
        pd.concat(
            [data.isnull().any(), 
             data.isnull().sum(),
             data.dtypes], 
            axis=1
        )
    )
    view_info.columns = ['Nulos', 'Cantidad', 'Tipo Col']
    return view_info

null_verificator(data)
```

```python
for col in ['NUM_PEDIDO', 'CODIGO_CONGLOMERADO_CLIENTE', 'CODIGO_CARACTERSTICA']:
    print("/////////////////// ", col)
    print(col, data[col].value_counts(dropna=False))
```

```python
#Debido a que el estudio se basa en los conglomerados
#Se eliminan los conglomerados nulos
data = data[data.CODIGO_CONGLOMERADO_CLIENTE.notnull()]
data.shape
```

```python
null_verificator(data)
```

```python
#Se reemplazara los datos nulos de ambas columnas por el valor 'desconocido'
#Aprovechando son de tipo object 
data.fillna('desconocido', inplace=True)
```

```python
import numpy as np
def reduce_size_data(df, save=True, new_name='data_qroma_train_reduce.csv'):
    for col in df.select_dtypes(include=['int']).columns:
        df[col] = df[col].apply(pd.to_numeric, downcast='integer')
 
    for col in df.select_dtypes(include=['float']).columns:
        df[col] = df[col].apply(pd.to_numeric, downcast='float')

    for col in df.select_dtypes(include=['object']).columns:
        if len(df[col].unique()) / len(df[col]) < 0.5:
            df[col] = df[col].astype('category')

    df.to_csv(new_name, index=False)
    
reduce_size_data(data)
get_memory_usage(data)
```

```python
def add_col_dates(data, col, format_match="%d-%b-%y", month=True, day=True, month_day=True,year=True,
                   weekday=True, replace_str=False, format_str_replace='%Y/%m/%d', replicate_in_test=False):
        """
        por optimizar en casos separados para data y test_data
        """
        data['date'] = pd.to_datetime(data[col], format=format_match)
        
        if month:
            print("Agregaraa mes")
            data['month'] = pd.to_numeric(data['date'].dt.strftime('%m'), errors='coerce')
            data['month'].fillna(-99)  
            
        if day:
            print("Agregaraa dia")
            data['day'] = pd.to_numeric(data['date'].dt.strftime('%d'), errors='coerce')
            data['day'].fillna(-99)
            
        if year:
            print("Agregaraa aÃ±o")
            data['year'] = pd.to_numeric(data['date'].dt.strftime('%Y'), errors='coerce')
            data['year'].fillna(-99)

        if month_day:
            print("Agregaraa mes dia")
            data['month_day'] = pd.to_numeric(data['date'].dt.strftime('%m%d'), errors='coerce')
            data['month_day'].fillna(-99)
           
        if weekday:
            print("Agregaraa dia de semana")
            data['weekday'] = pd.to_numeric(data['date'].dt.strftime('%w'), errors='coerce')
            data['weekday'].fillna(-99)
   
        if replace_str:
            data['date'] = data['date'].dt.strftime(format_str_replace)
    
add_col_dates(data, 'FECHA_BOLETO', format_match="%Y-%m-%d")
data.head(10)
```
