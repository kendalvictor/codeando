```python
curriculums['word_count'] = graphlab.text_analytics.count_words(curriculums['new_job_text'])
```

```python
curriculums.head()
```

```python
tfidf = graphlab.text_analytics.tf_idf(curriculums['word_count'])
```

```python
tfidf
```

```python
alenjandro = curriculums[curriculums['nombre_documento'] == 'alejand']
curriculums['tfidf'] = tfidf
```

```python
alenjandro[['tfidf']].stack('tfidf', new_column_name=['word', 'tfidf']).sort('tfidf', ascending=False)
```

```python
knn_model = graphlab.nearest_neighbors.create(curriculums, features=['tfidf'], label='nombre_documento')
```

```python
knn_model.query(alenjandro)
```
