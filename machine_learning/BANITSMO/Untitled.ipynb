```python
import pandas as pd
cols = ['v_{}'.format(_) for _ in range(0,13)]
data = pd.read_csv('hackaton_training_v1.csv', index_col=['v_0'], usecols=cols)
del data.index.name
data.head()
```

```python
def null_verificator(data):        
    if data.isnull().any().any():
        view_info = pd.DataFrame(
            pd.concat(
                [data.isnull().any(), 
                 data.isnull().sum(),
                 data.dtypes], 
                axis=1)
        )
        view_info.columns = ['Nulos', 'Cantidad', 'Tipo Col']
        return view_info
    else:
        return "DATA LIMPIA DE NULOS"
    
null_verificator(data)
```
