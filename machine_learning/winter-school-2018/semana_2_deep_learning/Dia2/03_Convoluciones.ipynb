```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
%matplotlib inline
```

# Cargamos una imagen

```python
img = plt.imread('imgs/dog2.jpg') / 255
print(img.shape)
plt.imshow(img);
```

```python
# Cambiamos el formato de la imagen a grises
gray = img.mean(axis=2)
print(gray.shape)
plt.imshow(gray, cmap='gray');
```

# Convoluciones

```python
kernel = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
])
```

```python
result = convolve2d(gray, kernel)
plt.imshow(result, cmap='gray');
```

```python
kernels = np.array([
    # Identity
    [[0, 0, 0],
     [0, 1, 0],
     [0, 0, 0]],
    # Blur
    [[1, 2, 1],
     [2, 4, 2],
     [1, 2, 1]],
    # Sharpen
    [[0, -1, 0],
     [-1, 5, -1],
     [0, -1, 0]],
    # Vertical edges
    [[0, 0, 0],
     [0, 1, -1],
     [0, 0, 0]],
    # Horizontal edges
    [[0, -1, 0],
     [0, 1, 0],
     [0, 0, 0]],
    # Outline
    [[-1, -1, -1],
     [-1, 8, -1],
     [-1, -1, -1]]
])
kernels.shape
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14,16))
filter_names = ['Identity', 'Blur', 'Sharpen', 'Vertical edges', 'Horizontal edges', 'Outline']

for k, n, ax in zip(kernels, filter_names, axes.flatten()):
    ax.imshow(convolve2d(gray, k), cmap='gray')
    ax.set_title(n)
    ax.set_axis_off()
```
