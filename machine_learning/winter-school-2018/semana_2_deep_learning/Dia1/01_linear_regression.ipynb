# Linear Regression Example
A linear regression learning algorithm example using Keras library adapted from.

- Author: Taegyun Jeon (tgjeon)
- Project: https://github.com/tgjeon/Keras-Tutorials

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate dataset
trX = np.asarray([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
                         7.042,10.791,5.313,7.997,5.654,9.27,3.1])
trY = np.asarray([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
                         2.827,3.465,1.65,2.904,2.42,2.94,1.3])
# create a y value which is approximately linear but with some random noise
```

```python
# Linear regression model
model = Sequential()
model.add(Dense(input_dim=1, output_dim=1, init='random_uniform', activation='linear')) # before init='uniform'
model.compile(optimizer='sgd', loss='mse')
```

```python
# Print initial weights
weights = model.layers[0].get_weights()
w_init = weights[0][0][0]
b_init = weights[1][0]
print('Linear regression model is initialized with weight w: %.2f, b: %.2f' % (w_init, b_init))
```

```python
# Train
model.fit(trX, trY, nb_epoch=1000, verbose=1)
```

```python
# Print trained weights
weights = model.layers[0].get_weights()
w = weights[0][0][0]
b = weights[1][0]
print('Linear regression model is trained with weight w: %.2f, b: %.2f' % (w, b))
```

```python
plt.plot(trX, trY, 'ro', label='Datos originales')
#plt.plot(trX, w_init*trX + b_init, label='init')
plt.plot(trX, w*trX + b, label='Predicción')
plt.legend()
plt.show()
```
