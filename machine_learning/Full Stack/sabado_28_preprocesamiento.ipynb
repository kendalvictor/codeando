```python
#Carga de datos
import pandas as pd
data = pd.read_csv('DS_Compra.csv', encoding= "ISO-8859-1", sep=",")
data.head()
```

```python
data.describe(include="all")
```

```python
def genera_rangos(valor):
    return '18-30' if valor <=30 else ('31-50' if valor <= 50 else ('50 a mas' if not pd.isnull(valor) else 'NULO'))
data['EDAD-RANGO'] = data['EDAD'].apply(lambda x: genera_rangos(x))
data.head()
```

```python
#Esto reeemplazara los nulos por su media
data['EDAD'].fillna(data['EDAD'].mean(), inplace=True)
data['ESTADOCIVIL'].fillna(data['ESTADOCIVIL'].mode()[0], inplace=True)
```

```python
data.isnull().any()
```

```python
import numpy as np
np.nanpercentile(data['EDAD'], [0,1,5,10,20,30,50,75,90,95,99,100])
```

```python
indice_percen_95 = data['EDAD'] > np.nanpercentile(data['EDAD'], [90])[0]
data.loc[indice_percen_95, "EDAD"] = np.nanpercentile(data['EDAD'], [90])[0]
```

```python
data.head()
```

```python
new_data = data.copy() #Juego de backup   new_data = data.copy()
```

```python
dicc_sexo = {'M': 1, 'F': 0}
data['SEXO'] = data['SEXO'].map(dicc_sexo)
data.head(10)
```

```python
estado_civil = pd.get_dummies(data['ESTADOCIVIL'], prefix='ESTADO', drop_first=True)
del data['ESTADOCIVIL']
data = pd.concat([data, estado_civil], axis=1)
data.head()
```

```python
dicc_rango = {
    '18-30': 1,
    '31-50': 2,
    '50 a mas': 0,
    'NULO': 3
}
data['EDAD-RANGO'] = data['EDAD-RANGO'].map(dicc_rango)
data.head()
```

```python
data = data.drop(['ID', 'FLAG_COMPRA'], axis=1)
data.head()
```

```python
recup_flag = pd.DataFrame(new_data['FLAG_COMPRA'])
data = pd.concat([data,recup_flag], axis=1)
data.head()
```

```python
data.to_csv("CLEAN_Compra.csv", index=False)
```
