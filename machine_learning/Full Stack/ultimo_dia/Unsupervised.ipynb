```python
import pandas as pd
data = pd.read_csv('College_Data.csv', index_col=0)
data_2 = data.copy()
data_2.head()
```

```python
data = data.drop('Private', axis=1)
data.head()
```

```python
data.describe(include="all")
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, random_state=1)
km.fit(data)
```

```python
km.cluster_centers_
```

```python
def converter(val):
    if val.lower() == 'yes':
        return 1
    return 0

data_2['cluster'] = data_2['Private'].apply(converter)
data_2.head()
```

```python
data_2['cluster_km'] = km.labels_
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(data_2['cluster'], data_2['cluster_km'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
kms = KMeans(n_clusters=2, random_state=1356677)
kms.fit(scaled_data)
```

```python
data_2['cluster_kms'] = kms.labels_
```

```python
confusion_matrix(data_2['cluster'], data_2['cluster_kms'])
```
