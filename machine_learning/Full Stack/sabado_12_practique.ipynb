```python
def verificator_null(data):
        if data.isnull().any().any():
            print(pd.concat(
                [data.isnull().any(), 
                 data.isnull().sum(),
                 data.dtypes], 
                axis=1))
        else:
            print("DATA LIMPIA DE NULOS")
verificator_null(data)
```
