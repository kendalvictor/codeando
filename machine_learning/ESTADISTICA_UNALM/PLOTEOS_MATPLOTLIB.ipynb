```python
import numpy as np
#import matplotlib
#matplotlib.use('Agg')

import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680806)

def plot_range(range_select, texto='Probando colocar texto'):
    fig, ax = plt.subplots()
    ax.plot(range_select, '-o', ms=15, lw=2, alpha=0.6, mfc='black')
    ax.grid()

    # position bottom right
    fig.text(0.8, -0.09, texto,
             fontsize=26, color='gray',
             ha='right', va='bottom', alpha=0.5)
    plt.show()
```

```python
plot_range(np.random.rand(20))
```

```python
plot_range(np.linspace(0, 100, 10))
```

```python
plot_range(np.logspace(1, 2, 10))
```

```python
from matplotlib.widgets import Slider, Button, RadioButtons, Cursor
fig, ax = plt.subplots()
rng = np.random.RandomState()
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 500 * rng.randn(100)
plt.scatter(x,y, c=colors, s=sizes, alpha=0.7)

plt.colorbar()
cursor = Cursor(ax, useblit=True, color='red', linewidth=90)
plt.show()
```
