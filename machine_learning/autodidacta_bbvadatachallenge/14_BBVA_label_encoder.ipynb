```python
from sklearn.neural_network import MLPClassifier
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
# graficos incrustados
%matplotlib inline

#Carga de los CSV
BBVA_data = pd.read_csv('train_clientes.csv', header=0)
print("Data Train: ", BBVA_data.shape)
BBVA_test = pd.read_csv('test_clientes.csv', header=0)
print("Data Test:  ", BBVA_test.shape)
```

```python
def completar_data(BBVA):
    if BBVA.isnull().any().any():
        tipos = BBVA.columns.to_series().groupby(BBVA.dtypes).groups
        ctext = tipos[np.dtype('object')]
        columnas = BBVA.columns  # lista de todas las columnas
        cnum = list(set(columnas) - set(ctext))
        #COmpletando los valores nulos de colummnas numericas con la media
        for c in cnum:
            BBVA[c] = BBVA[c].fillna(-5)
        # Completando valores nulos de columnas categ√≥ricas con la moda
        for c in ctext:
            BBVA[c] = BBVA[c].fillna('Desconocido')
        print("Clean Data Test: ", BBVA.shape, BBVA.isnull().any().any())

completar_data(BBVA_data)
completar_data(BBVA_test)
```

```python
from sklearn import preprocessing

EDAD_RANGE = pd.cut(BBVA_data['EDAD'], range(-5, 150, 10))
EDAD2, _ = pd.factorize(EDAD_RANGE)
fit_edad = preprocessing.LabelEncoder().fit(EDAD2)

ANTIGUEDAD_RANGE = pd.cut(BBVA_data['ANTIGUEDAD'], range(-2 ,20 , 2))
ANTIGUEDAD2, _ = pd.factorize(ANTIGUEDAD_RANGE)
fit_antiguedad = preprocessing.LabelEncoder().fit(ANTIGUEDAD2)

fit_ingreso = preprocessing.LabelEncoder().fit(BBVA_data['RANG_INGRESO'])
fit_flag = preprocessing.LabelEncoder().fit(BBVA_data['FLAG_LIMA_PROVINCIA'])
fit_pasivo = preprocessing.LabelEncoder().fit(BBVA_data['RANG_SDO_PASIVO_MENOS0'])
fit_producto = preprocessing.LabelEncoder().fit(BBVA_data['RANG_NRO_PRODUCTOS_MENOS0'])
```

```python
def new_factorizar_data(BBVA):
    EDAD_RANGE = pd.cut(BBVA['EDAD'], range(-5, 150, 10))
    EDAD2, _ = pd.factorize(EDAD_RANGE)
    BBVA['EDAD'] = fit_edad.transform(EDAD2)
    
    ANTIGUEDAD_RANGE = pd.cut(BBVA['ANTIGUEDAD'], range(-2 ,20 , 2))
    ANTIGUEDAD2, _ = pd.factorize(ANTIGUEDAD_RANGE)
    BBVA['ANTIGUEDAD'] = fit_antiguedad.transform(ANTIGUEDAD2)
    
    BBVA['RANG_INGRESO'] = fit_ingreso.transform(BBVA['RANG_INGRESO'])
    BBVA['FLAG_LIMA_PROVINCIA'] = fit_flag.transform(BBVA['FLAG_LIMA_PROVINCIA'])
    BBVA['RANG_SDO_PASIVO_MENOS0'] = fit_pasivo.transform(BBVA['RANG_SDO_PASIVO_MENOS0'])
    BBVA['RANG_NRO_PRODUCTOS_MENOS0'] = fit_producto.transform(BBVA['RANG_NRO_PRODUCTOS_MENOS0'])

new_factorizar_data(BBVA_data)
new_factorizar_data(BBVA_test)  
    
```
