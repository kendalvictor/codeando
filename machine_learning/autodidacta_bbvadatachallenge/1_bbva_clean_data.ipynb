```python
#Importaciones basicas
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
# graficos incrustados
%matplotlib inline
```

```python
BBVA_data = pd.read_csv('train_clientes.csv', header=0)
BBVA_data.shape
```

```python
BBVA_data.ix[:100, :54]
```

```python
BBVA_data.isnull().any()
```

```python
print("tipos:: ", BBVA_data.dtypes)
tipos = BBVA_data.columns.to_series().groupby(BBVA_data.dtypes).groups
tipos
```

```python
ctext = tipos[np.dtype('object')]
print(len(ctext))
print(ctext)
```

```python
columnas = BBVA_data.columns  # lista de todas las columnas
cnum = list(set(columnas) - set(ctext))
print(len(cnum))
print(cnum)
```

```python
#COmpletando los valores nulos de colummnas numericas con la media
for c in cnum:
    mean = BBVA_data[c].mean()
    BBVA_data[c] = BBVA_data[c].fillna(mean)
```

```python
# Completando valores nulos de columnas categóricas con la moda
for c in ctext:
    mode = BBVA_data[c].mode()[0]
    BBVA_data[c] = BBVA_data[c].fillna(mode)
```

```python
# Controlando que no hayan valores faltantes
BBVA_data.isnull().any().any()
```

```python
# Guardando el dataset preprocesado
BBVA_data.to_csv("clientes_procesado.csv", index=False)
```

```python
# Calculando el porcentaje de donantes sobre toda la base de datos
porcentaje_artitrion = (
    BBVA_data[BBVA_data.ATTRITION > 0]['ATTRITION'].count() * 1.0
       / BBVA_data['ATTRITION'].count()
) * 100.0
print("El procentaje de ARTITRION de la base de datos es {0:.2f}%".format(porcentaje_artitrion))
```
